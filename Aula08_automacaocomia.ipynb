Automação de arquivos

Com certeza! Aqui está um programa Python que organiza arquivos em pastas com base em suas extensões, usando as bibliotecas os e shutil. Ele irá criar pastas para cada tipo de arquivo e mover os arquivos para lá.

```python
import os
import shutil

# --- Configuração ----
# Define o diretório que você deseja organizar.
# Por padrão, ele usará o diretório atual onde o script está sendo executado.
# Você pode alterá-lo para um caminho específico, por exemplo: diretorio_alvo = '/caminho/para/seus/arquivos'
diretorio_alvo = os.getcwd()

print(f"Organizando arquivos no diretório: {diretorio_alvo}")

# --- Lógica de Organização ---
# Itera sobre todos os itens (arquivos e pastas) no diretório alvo.
for nome_item in os.listdir(diretorio_alvo):
    # Constrói o caminho completo para o item.
    caminho_item = os.path.join(diretorio_alvo, nome_item)

    # Verifica se o item é um arquivo (e não um diretório ou o próprio script).
    if os.path.isfile(caminho_item) and nome_item != os.path.basename(__file__):
        # Extrai a extensão do arquivo (ex: '.pdf', '.jpg').
        # os.path.splitext retorna uma tupla: (nome_base, extensão).
        _, extensao = os.path.splitext(nome_item)

        # Remove o ponto inicial da extensão para usar como nome da pasta.
        # Se não houver extensão (ex: 'arquivo_sem_extensao'), ele será tratado como 'outros'.
        extensao = extensao[1:].lower() # Converte para minúsculas para padronização

        # Se não houver extensão, categorize como 'outros'.
        if not extensao:
            nome_pasta = 'outros'
        else:
            nome_pasta = extensao

        # Constrói o caminho completo para a nova pasta de destino.
        caminho_pasta_destino = os.path.join(diretorio_alvo, nome_pasta)

        # Cria a pasta de destino se ela ainda não existir.
        # 'exist_ok=True' evita um erro se a pasta já existir.
        os.makedirs(caminho_pasta_destino, exist_ok=True)

        # Move o arquivo para a pasta de destino.
        # shutil.move lida com a movimentação e, se o destino for um diretório existente,
        # move o arquivo para dentro dele.
        try:
            shutil.move(caminho_item, caminho_pasta_destino)
            print(f"Movido: {nome_item} -> {nome_pasta}/")
        except Exception as e:
            print(f"Erro ao mover {nome_item}: {e}")

print("Organização de arquivos concluída!")
```

Consultas simples a APIs


```python
import requests # Importa a biblioteca 'requests' para fazer requisições HTTP a APIs.

def consultar_cep(cep): # Define a função 'consultar_cep' que recebe um CEP como argumento.

    # Limpa o CEP: remove hífens e pontos, e espaços em branco no início/fim.
    cep = cep.replace("-", "").replace(".", "").strip()

    # Verifica se o CEP tem 8 dígitos após a limpeza.
    if len(cep) != 8:
        return "CEP inválido." # Retorna uma mensagem de erro se o CEP não for válido.

    # Monta a URL da API ViaCEP com o CEP fornecido.
    url = f"https://viacep.com.br/ws/{cep}/json/"

    try: # Inicia um bloco try-except para tratar possíveis erros de conexão ou requisição.
        response = requests.get(url) # Faz uma requisição GET para a URL da API.

        response.raise_for_status() # Lança uma exceção HTTPError para respostas de status de erro (4xx ou 5xx).

        dados = response.json() # Converte a resposta JSON da API em um dicionário Python.

        # Verifica se a resposta JSON contém a chave 'erro', indicando que o CEP não foi encontrado.
        if "erro" in dados:
            return "CEP não encontrado na base de dados." # Retorna uma mensagem se o CEP não for encontrado.

        # Retorna as informações do logradouro, bairro e cidade formatadas.
        return f"Logradouro: {dados['logradouro']}, Bairro: {dados['bairro']}, Cidade: {dados['localidade']}"

    except requests.exceptions.RequestException as e: # Captura exceções específicas da biblioteca requests.
        return f"Erro de conexão: {e}" # Retorna uma mensagem de erro de conexão.

# Solicita ao usuário que digite o CEP que deseja consultar.
cep_digitado = input("Digite o CEP que deseja consultar: ")
# Chama a função 'consultar_cep' com o CEP digitado e imprime o resultado.
print(consultar_cep(cep_digitado))
```

Sistema de notificação

```python
import datetime # Importa o módulo datetime para trabalhar com datas e horas.

def enviar_notificacao(mensagem, tipo="INFO"): # Define a função 'enviar_notificacao' que aceita uma mensagem e um tipo de notificação (padrão é 'INFO').
    """
    Envia uma notificação com uma mensagem e um tipo (INFO, ALERTA, ERRO).
    Em um cenário real, esta função poderia enviar e-mails, SMS, ou usar APIs de serviços de notificação.
    """
    timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S") # Obtém a data e hora atuais e formata como uma string.
    if tipo.upper() == "ALERTA": # Verifica se o tipo de notificação é 'ALERTA' (ignorando maiúsculas/minúsculas).
        print(f"[{timestamp}] [ALERTA] {mensagem}") # Imprime a notificação de alerta com timestamp.
    elif tipo.upper() == "ERRO": # Se não for 'ALERTA', verifica se é 'ERRO'.
        print(f"[{timestamp}] [ERRO] {mensagem}") # Imprime a notificação de erro com timestamp.
    else: # Se não for 'ALERTA' nem 'ERRO', assume que é 'INFO' ou outro tipo padrão.
        print(f"[{timestamp}] [INFO] {mensagem}") # Imprime a notificação de informação com timestamp.

# --- Exemplos de Uso ---
print("\n--- Testando o Sistema de Notificação ---") # Imprime um cabeçalho para a seção de testes.

# Notificação de informação
enviar_notificacao("Operação concluída com sucesso.", "INFO") # Chama a função para enviar uma notificação de informação.

# Notificação de alerta
enviar_notificacao("Atenção: Espaço em disco baixo.", "ALERTA") # Chama a função para enviar uma notificação de alerta.

# Notificação de erro
enviar_notificacao("Erro crítico: Conexão com o banco de dados falhou.", "ERRO") # Chama a função para enviar uma notificação de erro.

# Exemplo de uso com uma condição (similar ao sistema de temperatura)
def verificar_status_servico(status_ok): # Define uma função para verificar o status de um serviço com base em um booleano.
    if status_ok: # Se o status for verdadeiro (True).
        enviar_notificacao("Serviço principal online.", "INFO") # Envia uma notificação de informação.
    else: # Se o status for falso (False).
        enviar_notificacao("Serviço principal offline! Investigar imediatamente.", "ERRO") # Envia uma notificação de erro.

print("\n--- Verificação de Status do Serviço ---") # Imprime um cabeçalho para a verificação de status.
verificar_status_servico(True) # Testa a função com status OK.
verificar_status_servico(False) # Testa a função com status não OK.

print("Sistema de notificação pronto.") # Imprime uma mensagem indicando o fim da execução.
```
