Número primo

```python
def calculadora():
    try:
        num1 = float(input("Digite o primeiro número: "))
        operador = input("Operação (+, -, *, /): ")
        num2 = float(input("Digite o segundo número: "))

        if operador == '+':
            resultado = num1 + num2
        elif operador == '-':
            resultado = num1 - num2
        elif operador == '*':
            resultado = num1 * num2
        elif operador == '/':
            # Tratamento de erro específico: Divisão por zero
            if num2 == 0:
                return "Erro: Não é possível dividir por zero."
            resultado = num1 / num2
        else:
            return "Operador inválido."

        return f"Resultado: {resultado}"
    except ValueError:
        return "Erro: Entrada inválida. Digite apenas números."

print(calculadora())
```

Inversão de string

```python
import random

def jogar_forca():
    # Lista de palavras para o jogo
    palavras = ['desenvolvimento', 'algoritmo', 'interface', 'software', 'python']
    palavra_secreta = random.choice(palavras).lower()

    # Estado inicial do jogo
    letras_descobertas = ['_' for _ in palavra_secreta]
    letras_tentadas = []
    tentativas = 6

    print("=== JOGO DA FORCA (Nível Intermediário) ===")

    while tentativas > 0 and '_' in letras_descobertas:
        print(f"\nPalavra: {' '.join(letras_descobertas)}")
        print(f"Tentativas restantes: {tentativas}")
        print(f"Letras já tentadas: {', '.join(letras_tentadas)}")

        chute = input("Chute uma letra: ").lower().strip()

        # Validação básica de entrada
        if len(chute) != 1 or not chute.isalpha():
            print("Por favor, digite apenas uma letra.")
            continue

        if chute in letras_tentadas:
            print(f"Você já tentou a letra '{chute}'. Tente outra!")
            continue

        letras_tentadas.append(chute)

        if chute in palavra_secreta:
            print(f"Boa! A letra '{chute}' está na palavra.")
            # Atualiza as posições da letra descoberta
            for i, letra in enumerate(palavra_secreta):
                if letra == chute:
                    letras_descobertas[i] = chute
        else:
            tentativas -= 1
            print(f"Que pena! A letra '{chute}' não existe na palavra.")

    # Finalização do jogo
    if '_' not in letras_descobertas:
        print(f"\nParabéns! Você venceu. A palavra era: {palavra_secreta}")
    else:
        print(f"\nGame Over! As tentativas acabaram. A palavra era: {palavra_secreta}")

if __name__ == "__main__":
    jogar_forca()
```
